## Male Female Ratio in Super Markets
* we can see that the number of male and female entering the store is almost equal.

```python
fig, ax = plt.subplots(figsize=(10,6))
ax = sns.countplot(x='Gender', data = df, palette='Set1')
ax.set_title("Gender Count in Super Market",fontsize=20)
ax.set_ylabel('Gender Count');
```

```python
df.groupby(['Customer type']).agg({'Total':'sum'})
```

```python
fig, ax = plt.subplots(figsize=(10,6))
ax = sns.countplot(x='Customer type', data = df, palette='Set2')
ax.set_title("Type of customers",fontsize=20)
ax.set_ylabel('Customer Count');
```

## Customer Types in Different Branches 
 
* There are almost same number of people in both categories
* Branch A has comparatively more number of normal people so they had to attract them by some offers :)

```python
fig, ax = plt.subplots(figsize=(10,6))
sns.set_theme(style="darkgrid")
ax = sns.countplot(x = 'Customer type', hue='Branch', data=df, palette='Set1')
ax.set_title("Customer Types in Different Branches",fontsize=20)
ax.set_ylabel('Customer Count');
```

## Gender ratio in different Branches

* Yangaon and Mandalay are two cities where Men shops more than Female but Naypyitaw has relatively less ratio
* Where Naypyitaw city has more Female customers than Men

```python
fig, ax = plt.subplots(figsize=(10,6))
ax = sns.countplot(x='City', hue='Gender', data=df, palette='Set2')
ax.set_title("Gender ratio in Different cities", fontsize=20);
```

## Types of Products brought by Male and  Female 

* Female brought most of products related to Fashion Accessories 
* Men brought most of products on Health and Beauty

```python
fig, ax = plt.subplots(figsize=(10,6))
ax = sns.countplot(x='Gender', hue='Product line', data=df, palette='Set2')
ax.set_title("Types of products Bought by Male and Female", fontsize=20);
```

## Product Sales according to time

* We can see that at 14:00 sales are in peak stage

```python
df['Time'] = pd.to_datetime(df['Time'])
df['Hour'] = (df['Time']).dt.hour
plt.figure(figsize=(14,6))
plt.style.use('classic')
SalesTime = sns.lineplot(x="Hour", y ="Quantity", data = df)
plt.title("Product sales per Hour");
```

## Checking the different payment methods used.

* E-Wallet and Cash Users are more
* Keep some discounts for credit card users to attract them

```python
plt.figure(figsize = (14,6))
sns.set_theme(style='darkgrid')
ax = sns.countplot(x = "Payment", data = df)
ax.set_title(label = "Payment methods of customers ", fontsize= 25)
ax.set_xlabel(xlabel = "Payment method")
ax.set_ylabel(ylabel = " Customer Count");
```

## Ratings Vs Branches

### 

## Using Boxplot
* We can see that the average rating of branch A and C is more than seven and branch B is less than 7

```python
plt.figure(figsize=(14,6)) 
ax = sns.boxplot(x="Branch", y = "Rating" ,data =df, palette= "RdYlBu")
ax.set_title("Rating distribution between branches", fontsize = 25)
ax.set_xlabel(xlabel = "Branches")
ax.set_ylabel(ylabel = "Rating distribution")
```

## Using Boxen Plot

```python
plt.figure(figsize=(10,6))
plt.style.use('classic')
ax = sns.boxenplot(x = "Quantity", y = "Product line", data = df, palette='Set1')
ax.set_title(label = "Average sales of different lines of products", fontsize = 25)
ax.set_xlabel(xlabel = "Qunatity Sales",fontsize = 16)
ax.set_ylabel(ylabel = "Product Line", fontsize = 16);
```

## Product sales on the basis of gender


```python
plt.style.use('classic')
plt.figure(figsize = (14,6))
ax= sns.stripplot(y= "Product line", x = "Total", hue = "Gender", data = df)
ax.set_title(label = "Product sales on the basis of gender")
ax.set_xlabel(xlabel = " Total sales of products")
ax.set_ylabel(ylabel = "Product Line");
```
